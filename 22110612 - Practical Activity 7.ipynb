```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Apply the whitegrid style
sns.set_style("whitegrid")

# Example line plot
sns.lineplot(x="Yellow cards", y="Red cards", data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting the context to "talk" for larger plots suitable for presentations
sns.set_context("talk")

# Example line plot
sns.lineplot(x="Club", y="Position", data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example distribution plot for Offsides column using histplot
sns.histplot(df['Offsides'], kde=True)
plt.xlabel("Offsides")
plt.ylabel("Density")
plt.title("Distribution of Offsides")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example KDE plot for Clean sheets column
sns.kdeplot(data=df['Clean sheets'], shade=True)
plt.xlabel("Clean sheets")
plt.ylabel("Density")
plt.title("Kernel Density Estimation Plot for Clean sheets")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example KDE plot for Clean sheets column
sns.kdeplot(data=df['Clean sheets'], fill=True)
plt.xlabel("Clean sheets")
plt.ylabel("Density")
plt.title("Kernel Density Estimation Plot for Clean sheets")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example rugplot for Last man tackles column
sns.rugplot(data=df['Last man tackles'])
plt.xlabel("Last man tackles")
plt.ylabel("Density")
plt.title("Rug Plot for Last man tackles")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example boxplot for Age with respect to Position
sns.boxplot(x='Position', y='Age', data=df)
plt.xlabel("Position")
plt.ylabel("Age")
plt.title("Boxplot of Age with respect to Position")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example violinplot for Appearances with respect to Nationality
sns.violinplot(x='Nationality', y='Appearances', data=df)
plt.xlabel("Nationality")
plt.ylabel("Appearances")
plt.title("Violinplot of Appearances with respect to Nationality")
plt.xticks(rotation=90)  # Rotates x-axis labels for better readability
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example stripplot for Wins with respect to Club
sns.stripplot(x='Club', y='Wins', data=df, jitter=True)
plt.xlabel("Club")
plt.ylabel("Wins")
plt.title("Stripplot of Wins with respect to Club")
plt.xticks(rotation=90)  # Rotates x-axis labels for better readability
plt.show()
```
